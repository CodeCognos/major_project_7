```python
import pandas as pd
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, confusion_matrix
import matplotlib.pyplot as plt
import seaborn as sns

# Sample DataFrame (assuming df1_clean is already prepared)
threshold = 60
df1_cleaned['Automata_Weak'] = df1_cleaned['Automata'] < threshold
df1_cleaned['Domain_Weak'] = df1_cleaned['Domain Score'] < threshold

# Define features and target for the classification model
features = df1_cleaned[['Quant', 'Logical', 'CP', 'Domain Score', 'Automata', 'Automata fix', 'writex']]
target = df1_cleaned[['Automata_Weak', 'Domain_Weak']]

# Split data into train and test sets
X_train, X_test, y_train, y_test = train_test_split(features, target, test_size=0.3, random_state=42)

# Initialize RandomForestClassifier
model = RandomForestClassifier(random_state=42)

# Define the parameter grid for GridSearchCV
param_grid = {
    'n_estimators': [100, 200, 300],
    'max_depth': [None, 10, 20, 30],
    'min_samples_split': [2, 5, 10]
}

# Initialize GridSearchCV
grid_search = GridSearchCV(estimator=model,
                           param_grid=param_grid,
                           cv=5,  # Number of cross-validation folds
                           n_jobs=-1,  # Use all available cores
                           verbose=2)  # Increase verbosity for debugging

# Fit GridSearchCV
grid_search.fit(X_train, y_train)

# Best parameters and best score
print("Best parameters:", grid_search.best_params_)
print("Best score:", grid_search.best_score_)

# Train model with the best parameters
best_model = grid_search.best_estimator_
y_pred = best_model.predict(X_test)

# Print classification report for each target
for i, column in enumerate(target.columns):
    print(f"Classification report for {column}:")
    print(classification_report(y_test[column], y_pred[:, i]))

# Confusion matrix for one of the target variables (e.g., 'Automata_Weak')
conf_matrix = confusion_matrix(y_test['Automata_Weak'], y_pred[:, 0])
plt.figure(figsize=(8, 6))
sns.heatmap(conf_matrix, annot=True, fmt='d', cmap='Blues',
            xticklabels=['Not Weak', 'Weak'],
            yticklabels=['Not Weak', 'Weak'])
plt.title('Confusion Matrix for Automata_Weak')
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.show()

# Feature importance
importances = best_model.feature_importances_
feature_names = features.columns
indices = importances.argsort()[::-1]

# Plot feature importances
plt.figure(figsize=(10, 6))
plt.title('Feature Importances')
plt.bar(range(len(feature_names)), importances[indices], align='center')
plt.xticks(range(len(feature_names)), [feature_names[i] for i in indices], rotation=45)
plt.xlim([-1, len(feature_names)])
plt.show()
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.multioutput import MultiOutputClassifier
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
import matplotlib.pyplot as plt
import seaborn as sns

# Sample DataFrame (assuming df1_cleaned is already prepared)
threshold_1 = 70
threshold_2 = 700
df1_cleaned['Automata_Weak'] = df1_cleaned['Automata'] < threshold_1
df1_cleaned['Domain_Weak'] = df1_cleaned['Domain Score'] < threshold_2

# Define features and target for the classification model
features = df1_cleaned[['Quant', 'Logical', 'CP', 'Domain Score', 'Automata', 'Automata fix', 'writex']]
target = df1_cleaned[['Automata_Weak', 'Domain_Weak']]

# Split data into train and test sets
X_train, X_test, y_train, y_test = train_test_split(features, target, test_size=0.3, random_state=42)

# Initialize models
models = {
    'Random Forest': RandomForestClassifier(random_state=42),
    'Logistic Regression': LogisticRegression(max_iter=1000, random_state=42),
    'Support Vector Classifier': SVC(probability=True, random_state=42),
    'Gradient Boosting': GradientBoostingClassifier(random_state=42)
}

# Define parameter grids for GridSearchCV
param_grids = {
    'Random Forest': {
        'estimator__n_estimators': [100, 200, 300],
        'estimator__max_depth': [None, 10, 20, 30],
        'estimator__min_samples_split': [2, 5, 10]
    },
    'Logistic Regression': {
        'estimator__C': [0.1, 1, 10],
        'estimator__penalty': ['l2']
    },
    'Support Vector Classifier': {
        'estimator__C': [0.1, 1, 10],
        'estimator__kernel': ['linear', 'rbf']
    },
    'Gradient Boosting': {
        'estimator__n_estimators': [100, 200],
        'estimator__learning_rate': [0.01, 0.1],
        'estimator__max_depth': [3, 5]
    }
}

# Initialize dictionary to store accuracy scores
model_accuracies = {}

# Function to train and evaluate models
def train_and_evaluate(model_name, model, param_grid):
    print(f"\nTraining and evaluating {model_name}...")
    
    # Wrap the model with MultiOutputClassifier
    multi_target_model = MultiOutputClassifier(model, n_jobs=-1)
    
    # Hyperparameter tuning
    grid_search = GridSearchCV(multi_target_model, param_grid, cv=5, n_jobs=-1, verbose=2)
    grid_search.fit(X_train, y_train)
    
    best_model = grid_search.best_estimator_
    y_pred = best_model.predict(X_test)
    
    # Print classification reports and calculate accuracy
    print(f"Best parameters for {model_name}:", grid_search.best_params_)
    accuracies = []
    for i, column in enumerate(target.columns):
        print(f"\nClassification report for {column} using {model_name}:")
        print(classification_report(y_test[column], y_pred[:, i]))
        
        accuracy = accuracy_score(y_test[column], y_pred[:, i])
        accuracies.append(accuracy)
    
    # Store the average accuracy for the model
    model_accuracies[model_name] = sum(accuracies) / len(accuracies)
    
    # Confusion matrices for each target variable
    for i, column in enumerate(target.columns):
        conf_matrix = confusion_matrix(y_test[column], y_pred[:, i])
        plt.figure(figsize=(8, 6))
        sns.heatmap(conf_matrix, annot=True, fmt='d', cmap='Blues',
                    xticklabels=['Not Weak', 'Weak'],
                    yticklabels=['Not Weak', 'Weak'])
        plt.title(f'Confusion Matrix for {column} using {model_name}')
        plt.xlabel('Predicted')
        plt.ylabel('Actual')
        plt.show()

# Train and evaluate all models
for name, model in models.items():
    train_and_evaluate(name, model, param_grids[name])

# Plot the comparison of model accuracies
plt.figure(figsize=(10, 6))
sns.barplot(x=list(model_accuracies.keys()), y=list(model_accuracies.values()), palette='viridis')
plt.title('Comparison of Model Accuracies')
plt.xlabel('Model')
plt.ylabel('Average Accuracy')
plt.ylim(0, 1)
plt.show()
```
